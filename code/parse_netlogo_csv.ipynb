```python
import csv

output_dict = {
    'link_probability': [],
    'population': [],
    'threshold': [],
    'media_1': [],
    'media_2':[],
    'media_1_bias':[],
    'media_2_bias':[],
    'steps':[],
    'final_values':[]
}

csv_file='/Users/peterinnes/polarization/experiments/test_1_nlogo_out.csv'

with open(csv_file) as csv_file:
    final_values_row=0
    for row in csv.reader(csv_file, delimiter=','):
        if len(row) > 0:
            print(row[0])

            if row[0]=='link-probability':
                link_prob=[]
                for value in row[1:]:
                    link_prob.append(value)
                output_dict['link_probability']=link_prob
            if row[0]=='population':
                pop=[]
                for value in row[1:]:
                    pop.append(value)
                output_dict['population']=pop
            if row[0]=='threshold':
                thresh=[]
                for value in row[1:]:
                    thresh.append(value)
                output_dict['threshold']=thresh
            if row[0]=='media-1':
                med1=[]
                for value in row[1:]:
                    med1.append(value)
                output_dict['media_1']=med1
            if row[0]=='media-2':
                med2=[]
                for value in row[1:]:
                    med2.append(value)
                output_dict['media_2']=med2
            if row[0]=='media-1-bias':
                med1bias=[]
                for value in row[1:]:
                    med1bias.append(value)
                output_dict['media_1_bias']=med1bias
            if row[0]=='media-2-bias':
                med2bias=[]
                for value in row[1:]:
                    med2bias.append(value)
                output_dict['media_2_bias']=med2bias
            if row[0]=='[steps]':
                steps=[]
                for value in row[1:]:
                    steps.append(value)
                output_dict['steps']=steps
            if final_values_row==1:
                final_values_row=2
            elif final_values_row==2:
                final_vals=[]
                for values in row[1:]:
                    #print(values)
                    values=values.strip('[').strip(']').split(' ')
                    float_values=[float(value) for value in values]
                    #print(values)
                    final_vals.append(float_values)
                output_dict['final_values']=final_vals
        elif not final_values_row:
            final_values_row=1
        
```
